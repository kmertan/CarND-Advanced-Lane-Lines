#### This notebook reads in images of the road from a camera fixed at the center of a vehicle. It's goal is to produce polynomials which approximate the curvature of the lane lines and will help determine the appropriate turning angle of the vehicle.

```python
# Begin by reading in the calibration images

import numpy as np
import cv2
import glob
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the number of checkerboard corners we want in the x and y directions
nx = 9
ny = 6

# Set object points and image points
objp = np.zeros((nx * ny, 3), np.float32)
objp[:, :2] = np.mgrid[0:nx, 0:ny].T.reshape(-1, 2)

# Create vectors for the 3d points we'll collect for object points and 2d vectors we'll collect for image points
objpoints = []
imgpoints = []

images = glob.glob('camera_cal/calibration*.jpg')

for index, img_path in enumerate(images):
    
    cal_img = mpimg.imread(img_path)
    cal_img_gray = cv2.cvtColor(cal_img, cv2.COLOR_BGR2GRAY) # Convert to grayscale

    # Find the corners using cv2
    ret, corners = cv2.findChessboardCorners(cal_img_gray, (nx, ny), None)

    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners) 

cal_img = mpimg.imread('camera_cal/calibration5.jpg')
img_shape = (cal_img.shape[1], cal_img.shape[0])

ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_shape, None, None) 

cal_undist = cv2.undistort(cal_img, mtx, dist, None, mtx)

f, ax = plt.subplots(1, 2, figsize = (20, 16))
ax[0].imshow(cal_img)
ax[0].set_title("Original")
ax[1].imshow(cal_undist)
ax[1].set_title("Undistorted")

#cv2.imwrite('./writeup_images/calibration_dist.jpg', cv2.cvtColor(cal_img, cv2.COLOR_RGB2BGR))
#cv2.imwrite('./writeup_images/calibration_undist.jpg', cv2.cvtColor(cal_undist, cv2.COLOR_RGB2BGR))
```

## The results agree with the lesson. We will use the S channel color gradient and X direction gradient thresholding combined for the project's data pipeline.

```python
def threshold_s_x_color(undist, threshold = (50, 200), threshold_s = (90, 150)):
    hls = cv2.cvtColor(undist, cv2.COLOR_BGR2HLS)
    s_only = hls[:, :, 2]
    
    binary_s_x_color = np.zeros_like(s_only)
    
    sobelx_s = cv2.Sobel(s_only, cv2.CV_64F, 1, 0)
    sobelx_s_abs = np.absolute(sobelx_s)
    sobelx_s_scaled = np.uint8((255 * sobelx_s_abs)/np.max(sobelx_s_abs))

    binary_s_x_color[((sobelx_s_scaled > threshold[0]) & (sobelx_s_scaled < threshold[1])) | ((s_only > threshold_s[0]) & (s_only < threshold_s[1]))] = 1
    
    return binary_s_x_color
```

## Now we create histograms of white pixel counts in each "column" of the image. This will allow us to find it's left and right peaks which we will take to be the lane lines.

```python
def hist_and_fit(warped):

    hist = np.sum(warped[warped.shape[0]//2:, :], axis = 0)

    midpoint = hist.shape[0]//2
    leftx_base = np.argmax(hist[:midpoint])
    rightx_base = np.argmax(hist[midpoint:]) + midpoint

    #Set the number of windows and determine the appropriate height
    nwindows = 9
    window_height = np.int(warped.shape[1]/nwindows)

    nonzero = warped.nonzero()
    nonzero_x = np.array(nonzero[1])
    nonzero_y = np.array(nonzero[0])

    leftx_current = leftx_base
    rightx_current = rightx_base

    margin = 100
    minpix = 50

    out_img = np.dstack((warped, warped, warped)) * 255 #Creates an image to draw windows on
    left_line_inds = []
    right_line_inds = []

    #Cycle through the windows

    for window in range(nwindows):
        #First we have to find where the corners of our window will be. 
        #This means finding the two y coordinates and four x coordinates.

        win_y_low = warped.shape[0] - (window + 1) * window_height
        win_y_high = warped.shape[0] - (window) * window_height
        win_x_left_low = leftx_current - margin
        win_x_left_high = leftx_current + margin
        win_x_right_low = rightx_current - margin
        win_x_right_high = rightx_current + margin

        #Draw the windows on our output image

        cv2.rectangle(out_img, (win_x_left_low, win_y_low), (win_x_left_high, win_y_high), (0, 255, 0), 2)
        cv2.rectangle(out_img, (win_x_right_low, win_y_low), (win_x_right_high, win_y_high), (0, 255, 0), 2)

        #Find the non-zero pixels on the left and right windows

        good_left_inds = ((nonzero_y >= win_y_low) & (nonzero_y < win_y_high) & (nonzero_x >= win_x_left_low) & (nonzero_x < win_x_left_high)).nonzero()[0]
        good_right_inds = ((nonzero_y >= win_y_low) & (nonzero_y < win_y_high) & (nonzero_x >= win_x_right_low) & (nonzero_x < win_x_right_high)).nonzero()[0]

        #Add these to the list of previously discovered left and right indices from prior window cycles

        left_line_inds.append(good_left_inds)
        right_line_inds.append(good_right_inds)

        #Recenter your window if we found a sufficient number of pixels

        if len(good_left_inds) >= minpix:
            leftx_current = np.int(np.mean(nonzero_x[good_left_inds]))

        if len(good_right_inds) >= minpix:
            rightx_current = np.int(np.mean(nonzero_x[good_right_inds]))

    left_line_inds = np.concatenate(left_line_inds)
    right_line_inds = np.concatenate(right_line_inds)

    #Find the nonzero index positions

    left_x = nonzero_x[left_line_inds]
    left_y = nonzero_y[left_line_inds]
    right_x = nonzero_x[right_line_inds]
    right_y = nonzero_y[right_line_inds]

    return left_x, left_y, right_x, right_y, left_line_inds, right_line_inds
```

```python
left_x, left_y, right_x, right_y, left_line_inds, right_line_inds = hist_and_fit(warped)

hist = np.sum(warped[warped.shape[0]//2:, :], axis = 0)
plt.plot(hist)
```

```python
left_fit = np.polyfit(left_y, left_x, 2)
right_fit = np.polyfit(right_y, right_x, 2)

ploty = np.linspace(0, warped.shape[0]-1, warped.shape[0])
left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]

nonzero = warped.nonzero()
nonzero_x = np.array(nonzero[1])
nonzero_y = np.array(nonzero[0])

out_img = np.dstack((warped, warped, warped)) * 255
out_img[nonzero_y[left_line_inds], nonzero_x[left_line_inds]] = [255, 0, 0]
out_img[nonzero_y[right_line_inds], nonzero_x[right_line_inds]] = [0, 0, 255]
plt.imshow(out_img)
plt.plot(left_fitx, ploty, color='yellow')
plt.plot(right_fitx, ploty, color='yellow')
plt.xlim(0, 1280)
plt.ylim(720, 0)
plt.savefig('./writeup_images/lines_drawn.jpg')
```

## Define a function to calculate curvature in "real world" space

```python
ym_per_pix = 30/720
xm_per_pix = 3.7/700

def calculate_curvature(left_fitx, right_fitx, ploty):
    y_eval = np.max(ploty)
    
    left_fit_real = np.polyfit(ploty * ym_per_pix, left_fitx * xm_per_pix, 2)
    right_fit_real = np.polyfit(ploty * ym_per_pix, right_fitx * xm_per_pix, 2)
    
    left_curve_rad = ((1 + (2 * left_fit_real[0] * y_eval * ym_per_pix + left_fit_real[1])**2)**1.5) / np.absolute(2 * left_fit_real[0])
    right_curve_rad = ((1 + (2 * right_fit_real[0] * y_eval * ym_per_pix + right_fit_real[1])**2)**1.5) / np.absolute(2 * right_fit_real[0])
    
    return left_curve_rad, right_curve_rad
```

## Define a function to project lane curves back onto the original image

```python
def project_back_to_original(undist, warped, left_fitx, right_fitx, ploty):
    warp_zero = np.zeros_like(warped).astype(np.uint8)
    warp_color = np.dstack((warp_zero, warp_zero, warp_zero))
    
    # Reformat the points into something we can plot
    
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))
    
    # Draw the lines onto the blank image we created
    
    cv2.fillPoly(warp_color, np.int_([pts]), (0, 255, 0))
    
    newwarp = cv2.warpPerspective(warp_color, Minv, (undist.shape[1], undist.shape[0]))
    result = cv2.addWeighted(undist, 1, newwarp, .3, 0)
    result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
    
    return result
```

## Define a Line class to help keep track of lines throughout time

```python
class Line():
    def __init__(self):
        self.detected = False # Checks if the line was detected in the last iteration
        self.recent_xfitted = [] # Stores the x values for the last n fitted lines
        self.bestx = None # Averages the fitted values of the last n iterations
        self.best_fit = None # Averages the polynomial coefficients of the last n iterations
        self.current_fit = np.array([False])
        self.curvature_radius = None # Stores the curvature radius in meters
        self.line_base_pos = None # Stores the distance from the center in meters
        self.diffs = np.array([0, 0, 0], dtype = 'float') # Stores the difference of the polynomial coefficients
        self.allx = None # X values for detected line pixels
        self.ally = None # Y values for detected line pixels
```

## Since we're working with a video now, I'll define a function to take advantage of previous detected lines in the current frame

```python
def hist_and_fit_from_prior(warped, left_fit, right_fit):
    nonzero = warped.nonzero()
    nonzero_x = np.array(nonzero[1])
    nonzero_y = np.array(nonzero[0])
    margin = 100
    
    left_lane_inds = ((nonzero_x > (left_fit[0] * (nonzero_y**2) + left_fit[1]*nonzero_y + left_fit[2] - margin)) & (nonzero_x < (left_fit[0] * (nonzero_y**2) + left_fit[1] * nonzero_y + left_fit[2] + margin)))
    right_lane_inds = ((nonzero_x > (right_fit[0] * (nonzero_y**2) + right_fit[1]*nonzero_y + right_fit[2] - margin)) & (nonzero_x < (right_fit[0] * (nonzero_y**2) + right_fit[1] * nonzero_y + right_fit[2] + margin)))
    
    left_x = nonzero_x[left_lane_inds]
    left_y = nonzero_y[left_lane_inds]
    right_x = nonzero_x[right_lane_inds]
    right_y = nonzero_y[right_lane_inds]
    
    return left_x, left_y, right_x, right_y
```

## Here we combine all of our steps into a single function that simply takes in an image and outputs the lane lines, curvature and offset superimposed onto it

```python
def lane_finder(image):
        
    img = np.copy(image)
    img_shape = (img.shape[1], img.shape[0])

    ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_shape, None, None)  

    undist = cv2.undistort(img, mtx, dist, None, mtx)
    
    binary_s_x_color = threshold_s_x_color(undist)
    
    warped = cv2.warpPerspective(binary_s_x_color, M, (binary_s_x_color.shape[1], binary_s_x_color.shape[0]), flags = cv2.INTER_NEAREST)

    if((right_line.detected == False) or (left_line.detected == False)):
        left_x, left_y, right_x, right_y, left_line_inds, right_line_inds = hist_and_fit(warped) #left_line_inds and right_line_inds ignored. Only used for plotting purposes above.
        
    else:
        left_x, left_y, right_x, right_y = hist_and_fit_from_prior(warped, left_line.current_fit, right_line.current_fit)
    
    if((len(right_x) == 0) or (len(left_x) == 0)):
        left_fit = left_line.current_fit
        right_fit = right_line.current_fit
        
    else:
        left_fit = np.polyfit(left_y, left_x, 2)
        right_fit = np.polyfit(right_y, right_x, 2)
    
        left_line.current_fit = left_fit
        right_line.current_fit = right_fit
    
    ploty = np.linspace(0, warped.shape[0]-1, warped.shape[0])
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    
    left_curve, right_curve = calculate_curvature(left_fitx, right_fitx, ploty)
    
    distance_variance = (np.var(left_fitx - right_fitx))
    distance_mean = np.absolute(np.mean(left_fitx - right_fitx))
    
    if((distance_variance > 1000) or (distance_mean < 850)):
        right_fitx = right_line.recent_xfitted[-1]
        left_fitx = left_line.recent_xfitted[-1]
        right_line.detected = False
        left_line.detected = False
        
    else:
        right_line.detected = True
        left_line.detected = True        
        
    left_line.recent_xfitted.append(left_fitx)
    right_line.recent_xfitted.append(right_fitx)
    
    right_fitx = np.mean(right_line.recent_xfitted[-20:], axis = 0)
    left_fitx = np.mean(left_line.recent_xfitted[-20:], axis = 0)
    
    result = project_back_to_original(undist, warped, left_fitx, right_fitx, ploty)
    result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
    
    curvature = calculate_curvature(left_fitx, right_fitx, ploty)
    midpoint = np.mean((np.mean(left_fitx), np.mean(right_fitx)))
    offset = round(np.absolute(midpoint - img_shape[0]/2) * xm_per_pix, 2)
    
    avg_curvature = round((curvature[1] + curvature[0]) / 2, 2)
    cv2.putText(result, "Curvature: " + str(avg_curvature) + " m", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 0), 3)
    cv2.putText(result, "Offset: " + str(offset) + " m", (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 0), 3)
    
    return result
```
